# UC 3: Compare Two Lines (Greater / Smaller)

```python
import math

def line_lenght(x1,x2,y1,y2):
    return math.sqrt((x2-x1)**2 + (y2-y1)**2)

print("Enter the lenght of line1 : ")
x1 = float(input("Enter the x1 : "))
x2 = float(input("Enter the x2 : "))
y1 = float(input("Enter the y1 : "))
y2 = float(input("Enter the y2 : "))

print("Enter the lenght of line2 : ")
a1 = float(input("Enter the a1 : "))
a2 = float(input("Enter the a2 : "))
b1 = float(input("Enter the b1 : "))
b2 = float(input("Enter the b2 : "))

l1 = line_lenght(x1,x2,y1,y2)
l2 = line_lenght(a1,a2,b1,b2)

if l1 > l2:
    print("\nLine 1 is greater")
elif l1 < l2:
    print("\nLine 1 is smaller")
else:
    print("\nBoth lines are equal")
```
